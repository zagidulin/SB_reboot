```python
# task1  (lesson7)
# sqlite3: Сделать тестовый проект с БД (sqlite3, project name: task1_7).
# В таблицу table1 записать 1000 строк с случайными значениями (3 колонки, тип int) от 0 до 1000.
# Далее построить гистаграмму распределения этих трех колонко

conn = sqlite3.connect('task1_7')
c = conn.cursor()

request = "CREATE TABLE table1(col1 int, col2 int, col3 int)"
c.execute(request)

request = '''
INSERT INTO table1
with recursive generate_series(value) as (
    select 1
    union all select value+1
    from generate_series
    where value+1<=1000)
select abs(random() % 1000), abs(random() % 1000), abs(random() % 1000) from generate_series;
'''
c.execute(request)

```

```python
df = pd.read_sql("select * from table1;", con=conn)
```

```python
x0 = df['col1']
x1 = df['col2']
x2 = df['col3']

fig = go.Figure()
fig.add_trace(go.Histogram(
    x=x0,
    name='col1', # name used in legend and hover labels
    xbins=dict( # bins used for histogram
        start=0,
        end=1000,
        size=20
    ),
    marker_color='#EB89B5',
    opacity=0.75
))
fig.add_trace(go.Histogram(
    x=x1,
    name='col2',
    xbins=dict(
        start=0,
        end=1000,
        size=20
    ),
    marker_color='#330C73',
    opacity=0.75
))
fig.add_trace(go.Histogram(
    x=x2,
    name='col3',
    xbins=dict(
        start=0,
        end=1000,
        size=20
    ),
    marker_color='#770737',
    opacity=0.75
))

fig.update_layout(
    title_text='table1 data', # title of plot
    xaxis_title_text='Value', # xaxis label
    yaxis_title_text='Count', # yaxis label
    bargap=0.2, # gap between bars of adjacent location coordinates
    bargroupgap=0.1 # gap between bars of the same location coordinates
)

fig.show()
```

```python
## task2  (lesson7)
-- oracle: https://leetcode.com/problems/duplicate-emails/

with q as (select email, count(email) as freq
           from Person
           group by email)
select email from q
where freq > 1;
```

```python
## task3  (lesson7)
-- oracle: https://leetcode.com/problems/employees-earning-more-than-their-managers/

select Emp1.name as Employee from Employee Emp1
join Employee Emp2
on Emp2.id = Emp1.managerId
where Emp1.salary > Emp2.salary;
```
